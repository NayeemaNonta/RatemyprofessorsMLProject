#Importing and Setup

```python
# Import necessary python libraries
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# upload Rate My Professor data for exploration
from google.colab import files 
uploaded_files = files.upload() # upload file
```

```python
import io
data = io.BytesIO(uploaded_files['RateMyProfessor_Sample_data.csv']) # read file
```

```python
# load as a data frame
df = pd.read_csv(data) 

# view the first 5 rows to get an initial idea of the columns
df.head()
```

#Data Preprocessing: Step 1
Keeping only unique instances of the Professor Name column

```python
# output the count of all rows in the dataset before any modifications are made
print("Count before Removing Duplicates: ")
df.count()
```

```python
# output the first 5 rows of dataset 
df.head()

# Note: all Profs listed here the same Prof with the same 'star_rating' value
```

```python
# remove rows that have a duplicate Professor Name
ab = df.drop_duplicates(subset='professor_name', keep='first')

# output the count of all rows in the dataset after removing duplicates 
print("Count after Removing Duplicates: ")
ab.count()
```

```python
# output the first 5 rows of dataset 
ab.head()

# Note: all Profs listed here are different with different 'star_rating' values
```

#Data Preprocessing: Step 2
Dropping rows where there are nulls in "tag_professor" column. This is important because otherwise it will overinflate the '0' values for the 20 binary Professor Characteristics columns such that meaningful predictions may not be made. 

```python
# output the count of all rows in the dataset before any modifications are made
print("Count before dropping:\n", ab.count(), "\n")
```

```python
# dropping rows if there are missing values present in tag_professor
cd = ab.dropna(subset=['tag_professor'])
# output the count of all rows in the dataset after the modifications are made
print("Count after dropping:\n", cd.count())
```

#Data Preprocessing: Step 3
Developing 2 new datasets: Good Professors vs Bad Professors. 
Professors with a rating greater than 4.0 will be considered Good Professors and Professors with a Rating under 2.9 will be considered Bad Professors. All Professors with ratings 3.0 to 3.9 will be omitted.

```python
# output the first 5 rows of dataset 
cd.head()

# Note: 1 Prof (William Hollinrake) listed here has a star_rating of 1.6 making him a Bad Professor
# Note: The rest 4 Profs are considered Good Professors as their star_rating is above 4.0
```

```python
# bad profs are rated from 4.0-5.0 (star_rating column)
# finding values where the professors are rated under 4.0 are dropped 
goodProfs = cd.drop(cd[cd.star_rating < 4.0].index) 

# counting the number of Good Professors in the dataset
goodProfs.count() #283 rows of data
```

```python
# output the first 5 rows of Good Profs dataset 
goodProfs.head()

# Note: Now all Profs listed here has a star_rating greater than or equal to 4.0 making them Good Professors
```

```python
# bad profs are rated from 1.0-2.9 (star_rating column)
# finding values where the professors are rated above or equal to 3.0 and dropping them
badProfs = cd.drop(cd[cd.star_rating >= 3.0].index) 

# counting the number of Bad Professors in the dataset
badProfs.count() #99 rows of data
```

```python
# output the first 5 rows of Bad Profs dataset 
badProfs.head()

# Note: Now all Profs listed here has a star_rating less than or equal to 2.9 making them Good Professors
```

#Data Preprocessing: Step 4
Creating the final 2 datasets by only including the 20 boolean Characteristics for both the Good Profs and Bad Profs.

```python
# converting the goodProfs and badProfs datasets to CSV files for reference
goodProfs.to_csv('good_profs.csv')
badProfs.to_csv('bad_profs.csv')
```

```python
# including only the boolean columns of tags for good professor
goodProfsBooleanTags = goodProfs[['gives_good_feedback', 'caring', 'respected', 'participation_matters', 'clear_grading_criteria', 'skip_class', 'amazing_lectures', 'inspirational', 'tough_grader', 'hilarious', 'get_ready_to_read', 'lots_of_homework', 'accessible_outside_class', 'lecture_heavy', 'extra_credit', 'graded_by_few_things', 'group_projects', 'test_heavy', 'so_many_papers', 'beware_of_pop_quizzes']]
# converting the goodProfsBooleanTags to CSV files for reference
goodProfsBooleanTags.to_csv('good_profs_boolean_tags.csv')
```

```python
# output the first 5 rows of Good Profs boolean tags dataset 
goodProfsBooleanTags.head()
```

```python
# including only the boolean columns of tags for good professor
badProfsBooleanTags = badProfs[['gives_good_feedback', 'caring', 'respected', 'participation_matters', 'clear_grading_criteria', 'skip_class', 'amazing_lectures', 'inspirational', 'tough_grader', 'hilarious', 'get_ready_to_read', 'lots_of_homework', 'accessible_outside_class', 'lecture_heavy', 'extra_credit', 'graded_by_few_things', 'group_projects', 'test_heavy', 'so_many_papers', 'beware_of_pop_quizzes']]
# converting the badProfsBooleanTags to CSV files for reference
badProfsBooleanTags.to_csv('bad_profs_boolean_tags.csv')
```

```python
# output the first 5 rows of Good Profs boolean tags dataset 
badProfsBooleanTags.head()
```

# **Association Rules: Good Profs**
4 experiments will be done to understand interesting correlations between Characteristics of Good Professors

1. Low support and Low confidence
2. Low support and High confidence
3. High support and Low confidence
4. High support and High confidence



###*EXPERIMENT 1: Low support and Low confidence*

```python
# GOOD PROFS:
# find all itemsets that have a minimum support of 10%
frequent_itemsets_good = apriori(goodProfsBooleanTags,min_support=0.1,use_colnames=1)

# print the itemsets that meet the criteria
print(frequent_itemsets_good)
```

```python
# GOOD PROFS:
# find association rules that have confidence >= 10%
rules_good_prof = association_rules(frequent_itemsets_good,metric='confidence',min_threshold=0.1)

# print the association rules, their support, and confidence
print(rules_good_prof[['antecedents','consequents','support','confidence']])
```
